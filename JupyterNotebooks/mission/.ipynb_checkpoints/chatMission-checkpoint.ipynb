## Part D

rewrite the sysmlv2 code for this document by basing it off the syntax of this sysmlv2 system:

part def Camera {
	import PictureTaking::*;
	
	perform action takePicture[*] :> PictureTaking::takePicture;
	
	part focusingSubsystem {
		perform takePicture.focus;
	}
	
	part imagingSubsystem {
		perform takePicture.shoot;
	}
	
}

package PictureTaking {
	part def Exposure;
	
	action def Focus { out xrsl: Exposure; }
	action def Shoot { in xsf: Exposure; }	
		
	action takePicture {		
		action focus: Focus[1];
		flow focus.xrsl to shoot.xsf;
		action shoot: Shoot[1];
	}
}

```sysml
package DataLinks {
  part def DataLink {
    import UAS::*;
    import UGV::*;
    
    perform action exchangeData[*] :> UAS::exchangeData;
    perform action receiveData[*] :> UGV::receiveData;
  }
  
  package UAS {
    part def UASDataLink extends DataLink {
      perform exchangeData;
    }
  }
  
  package UGV {
    part def UGVDataLink extends DataLink {
      perform receiveData;
    }
  }
}

part def OpticSensor {
  import UGV::*;
  
  perform action detectLandmines[*] :> UGV::detectLandmines;
}

part def BaseSecurityControlCenter {
  action monitorSecurity[*];
}
```

```sysml
package DataLinks {
  part def DataLink {
    
    action def exchangeData {
        out Data;
    }
    
    action def receiveData {
        in Data;
    }
  }
  
  package UAS {
    part def UASDataLink {
      perform action exchangeData;
    }
  }
  
  package UGV {
    part def UGVDataLink {
      perform action receiveData;
    }
  }
}

part def OpticSensor {
   
  action def detectLandmines {
      out detectMines;
  }
}

part def BaseSecurityControlCenter {
  action def monitorSecurity {
      out monitor;
  }
}
```
