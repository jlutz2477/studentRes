```sysml
package UAS {
    import Base::*;

    part UAS {
        name: 'Unmanned Aerial System';

        // Vehicle subsystem
        part vehicle: Vehicle;

        // Control system
        part controlSystem: ControlSystem;

        // Sensor suite
        part sensors: Sensors;

        // Communication system
        part communications: Communications;
    }

    part definition Vehicle {
        name: 'Drone Vehicle';

        attribute weight: Mass;
        attribute maxSpeed: Speed;

        part propulsionSystem: PropulsionSystem;
        part battery: Battery;
    }

    part definition PropulsionSystem {
        name: 'Propulsion System';
        attribute type: String = "Quadcopter";
    }

    part definition Battery {
        name: 'Battery';
        attribute capacity: Energy;
        attribute voltage: Voltage;
    }

    part definition ControlSystem {
        name: 'Control System';
        port controlInput: ControlInput;
        port controlOutput: ControlOutput;
    }

    part definition Sensors {
        name: 'Sensor Suite';
        part gps: GPS;
        part camera: Camera;
    }

    part definition Communications {
        name: 'Communication System';
        attribute frequency: Frequency;
    }

    part definition GPS {
        name: 'GPS Receiver';
    }

    part definition Camera {
        name: 'Optical Camera';
    }

    part definition ControlInput {
        name: 'Control Input Interface';
    }

    part definition ControlOutput {
        name: 'Control Output Interface';
    }
}

package Base {
    datatype Mass {
        unit kg;
    }
    datatype Speed {
        unit m_per_s;
    }
    datatype Energy {
        unit kWh;
    }
    datatype Voltage {
        unit V;
    }
    datatype Frequency {
        unit Hz;
    }
}
```
