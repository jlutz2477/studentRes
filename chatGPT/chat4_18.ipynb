```sysml
package UnmannedGroundVehicle {
    part def UGV {
        part chassis: Chassis;
        part navSystem: NavigationSystem;
        part sensors: Sensors;
        part communication: Communication;

        stateMachine behavior = define UGVBehavior();
        action operate = define OperateUGV();
    }

    part def Chassis {
        attribute material: String;
        attribute weight: Real;
    }

    part def NavigationSystem {
        part gpsModule: GPSModule;
        part computingUnit: ComputingUnit;
    }

    part def Sensors {
        part camera: Camera;
        part lidar: LIDAR;
        part thermalSensor: ThermalSensor;
    }

    part def Communication {
        attribute frequency: Real;
        attribute bandwidth: Real;
    }

    stateMachine def UGVBehavior() {
        state initialState;
        state operational;
        state maintenance;
        state emergency;

        transition from initialState to operational;
        transition from operational to maintenance on condition(maintenanceRequired);
        transition from operational to emergency on condition(emergencyDetected);
        transition from emergency to operational on condition(emergencyResolved);
        transition from maintenance to operational on condition(maintenanceCompleted);
    }

    action def OperateUGV() {
        perform navigation = define Navigate();
        perform sensing = define GatherData();
        perform communicate = define TransmitData();
    }

    action def Navigate() {
        base Action;
        input from sensors.camera;
        input from sensors.lidar;
        output to navSystem.gpsModule;
    }

    action def GatherData() {
        base Action;
        input from sensors;
        output to navSystem.computingUnit;
    }

    action def TransmitData() {
        base Action;
        input from navSystem.computingUnit;
        output to communication;
    }

    part def GPSModule {
        attribute coordinates: String;
    }

    part def ComputingUnit {
        attribute processorSpeed: Real;
    }

    part def Camera {
        attribute resolution: Integer;
    }

    part def LIDAR {
        attribute range: Real;
    }

    part def ThermalSensor {
        attribute sensitivity: Real;
    }
}
```
